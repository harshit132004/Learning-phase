# BAR PLOT

```python
#Basic Syntax
"""df=sns.load_datadet("")
   sns.barplot(x="",y="",data=df)
   plt.show()"""
```

```python
data=sns.load_dataset("penguins")
```

```python
sns.barplot(x="island",y="bill_length_mm",hue="sex",data=data,palette=["g","r"],saturation=1,err_kws={"color": "y"}
           ,capsize=0.5,dodge=False)
                                                                                                    
plt.show()
```

```python
sns.set(style="ticks")
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island", y="bill_length_mm", data=data, hue="sex", hue_order=["Female", "Male"], order=order_1
           ,errorbar=('ci', 100),n_boot=3,saturation=30)#ci=100
plt.show()

```

```python
sns.barplot(x="flipper_length_mm", y="bill_length_mm", data=data, hue="sex", hue_order=["Female", "Male"],orient="h")
plt.show()
```

```python
var_1=np.linspace(1,10,10)
var_1=var_1.reshape(2,5)
var_1
```

```python
y={"fontsize":20,
  "color":"r"
  }
v=sns.heatmap(var_1,vmin=0,vmax=10,cmap="PuOr",annot=True,annot_kws=y,linewidth=4,linecolor="b",cbar=False,
           xticklabels=False,yticklabels=False)

v.set(xlabel="python",ylabel="java")
sns.set(font_scale=15)

plt.show()
```

# BOX PLOT

```python
#BASIC SYNTAX
"""df=sns.load_dataset("")
   sns.boxplot(x="",y="",data=df)
   plt.show()"""
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,palette="jet_r",hue="sex",order=["Sun","Thur","Fri","Sat"],showmeans=True)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(x="day",y="total_bill",data=df,palette="hot",showmeans=True,meanprops={"marker":"o",
                                                                                  "markeredgecolor":"k"},
           linewidth=3,orient="v")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="day",x="total_bill",data=df,palette="plasma",showmeans=True,meanprops={"marker":"o",
                                                                                  "markeredgecolor":"k"},
           linewidth=3)
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(data=df,palette="hot",showmeans=True,meanprops={"marker":"o",
                                                                                  "markeredgecolor":"k"},
           linewidth=3,orient="h")
plt.show()
```

```python
sns.set(style="whitegrid")
sns.boxplot(y="total_bill",data=df,color="r")
plt.show()
```
