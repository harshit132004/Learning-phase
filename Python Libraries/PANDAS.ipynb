```python
import pandas as pd

# Create the DataFrame
var_1 = pd.DataFrame({"A": [1, 2, 3, 4, 5], "B": [9, 8, 7, 6, 5]})

# Display the original DataFrame
print("Original DataFrame:")
print(var_1)

# Create a new column "python" with the first three elements of column "A"
var_1["python"] = var_1["A"][:3]

# Display the DataFrame after adding the new column
print("\nDataFrame after adding 'python' column:")
print(var_1)

# Remove the "python" column and store it in var_2
var_2 = var_1.pop("python")

# Display the updated DataFrame and the popped Series
print("\nUpdated DataFrame after popping 'python' column:")
print(var_1)
print("\nPopped 'python' column:")
print(var_2)
```

```python
import pandas as pd
import numpy as np

# Create a DataFrame with some NaN values
data = {
    "Category": ["A", np.nan, "C", np.nan, "E"],
    "Count": [10, np.nan, 30, np.nan, 50]
}
var = pd.DataFrame(data)

print("Original DataFrame:")
print(var)

# Convert object dtypes to inferred types
var = var.infer_objects(copy=False)

# Interpolate missing values
var.interpolate(method="linear", axis=0, inplace=True)

print("\nDataFrame after interpolation:")
print(var)
```

## MERGE

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,on="A",how="left")
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,on="A",how="right")
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,on="A",how="outer")
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,on="A",how="right",indicator=True)
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,how="right",indicator=True,left_index=True,right_index=True)
```

```python
var_1=pd.DataFrame({"A":[1,2,3,4],
                   "B":[5,6,7,8]})
var_2=pd.DataFrame({"A":[1,2,3,5],
                   "B":[15,16,17,18]})

pd.merge(var_1,var_2,how="right",indicator=True,left_index=True,right_index=True,suffixes=("name","python"))
```
