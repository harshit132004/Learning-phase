```python
temp_df=player.merge(player_match,on="Player_Id")[["Player_Name","Match_Id","Is_Captain"]]
```

```python
delivery.merge(temp_df,left_on=[ "match_id","batter"],right_on=["Match_Id","Player_Name"],how="left").fillna(0)
```

```python
runs=delivery.groupby(["match_id","batter"])["batsman_runs"].sum().reset_index()
balls_played=delivery.groupby(["match_id","batter"])["batsman_runs"].count().reset_index()
```

```python
balls_played
```

```python
sixs=delivery.query("batsman_runs==6").groupby(["match_id","batter"])["batsman_runs"].count().reset_index()
fours=delivery.query("batsman_runs==4").groupby(["match_id","batter"])["batsman_runs"].count().reset_index()
```

```python
fours
```

```python
final_df=runs.merge(balls_played, on=["match_id","batter"],suffixes=("_runs","_balls")).merge(fours,on=["match_id","batter"],how="left").merge(sixs,on=["match_id","batter"],how="left")
```

```python
final_df.fillna(0,inplace=True)
final_df.rename(columns={"batsman_runs_runs":"RUNS",
                            "batsman_runs_balls":"BALLS",
                           "batsman_runs_x":"SIXS",
                           "batsman_runs_y":"FOURS" },inplace=True)
```

```python
final_df["sr"]=round((final_df["RUNS"]/final_df["BALLS"])*100,2)
```

```python
final_df
```

```python
final_df.drop(columns=["BALLS"],inplace=True)
```

```python
final_df = final_df.merge(temp_df,left_on=['match_id','batter'],right_on=['Match_Id','Player_Name'],how='left').drop(columns=['Player_Name','Match_Id']).fillna(0)
```

```python
final_df
```

```python
final_df = final_df.merge(balls_played,on=['match_id','batter']).rename(columns={'batsman_run':'balls'})
```

```python
def dream11(row):
  score = 0

  score = score + row['RUNS'] + row['FOURS'] + 2*row['SIXS']

  if row['RUNS'] >= 100:
    score = score + 16

  elif row['RUNS'] >= 50 and row['RUNS'] < 100:
    score = score + 8

  elif row['RUNS'] >= 30 and row['RUNS'] < 50:
    score = score + 4

  elif row['RUNS'] == 0:
    score = score - 2

  if row['batsman_runs'] >= 10:

    if row['sr'] > 170:
      score = score + 6

    elif row['sr'] > 150 and row['sr'] <= 170:
      score = score + 4

    elif row['sr'] > 130 and row['sr'] <= 150:
      score = score + 2

    elif row['sr'] > 60 and row['sr'] <= 70:
      score = score - 2

    elif row['sr'] > 50 and row['sr'] <= 60:
      score = score - 4

    elif row['sr'] <= 50:
      score = score - 6
    else:
      pass

  if row['Is_Captain'] == 1:
    score = score*2


  return score
```

```python
final_df.apply(dream11,axis=1)
```

```python
final_df["score"]=final_df.apply(dream11,axis=1)
```

```python
final_df.sort_values( "score",ascending=False)
```
