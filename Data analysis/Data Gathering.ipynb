```python
import requests
import pandas as pd

url = "https://covid-193.p.rapidapi.com/statistics"

headers = {
    "x-rapidapi-key": "64cf44e786mshcc7550ccb4b37c3p1408acjsne3ee5d007544",
    "x-rapidapi-host": "covid-193.p.rapidapi.com"
}

response = requests.get(url, headers=headers)

# Convert the JSON response to a dictionary
data = response.json()

# Print the keys and a sample entry to understand the structure
data.keys()
(data['response'][0])  # Print the first entry in 'response'
```

```python
from pandas import json_normalize

# Normalize the JSON data
normalized_data = json_normalize(data['response'])

# Convert to DataFrame
df = pd.DataFrame(normalized_data)

# Display the DataFrame
df = pd.DataFrame(normalized_data)[["country","population","day","time","cases.new","cases.active",
                                    "deaths.total","tests.total"]]
df = pd.to_datetime(["day","time"])
df
```

```python
import pandas as pd
from pandas import json_normalize

# Normalize the JSON data
normalized_data = json_normalize(data['response'])

# Select specific columns
df = normalized_data[[
    "country",
    "population",
    "day",
    "time",
    "cases.new",
    "cases.active",
    "deaths.total",
    "tests.total"
]]

# Combine 'day' and 'time' into a single datetime column
df['datetime'] = pd.to_datetime(df['day'] + ' ' + df['time'])

# Drop the original 'day' and 'time' columns if no longer needed
df.drop(columns=['day', 'time'], inplace=True)

# Display the DataFrame with the new 'datetime' column
(df)
```
